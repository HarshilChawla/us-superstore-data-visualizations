```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv(r'C:\DATA ANALYSIS AND DATA SCIENCE PROJECTS\US Superstore Data Visualizations\data\us_superstore_dataset.csv')
```

# BAR GRAPH
*Each bar is a category, bar length is a value*

```python
data.columns
```

```python
#Sales by Region
total_sales = data.groupby('Region')['Sales'].sum().reset_index()
colordict = {'Central': 'red', 'East': 'yellow', 'South': 'blue', 'West': 'orange'}
colors = total_sales['Region'].map(colordict)
plt.figure(figsize = (8, 6))
bars = plt.bar(total_sales['Region'], total_sales['Sales'], 
        color = colors, 
        width = 0.8,
        edgecolor='black', 
        linewidth=1.5)
plt.title('Sales by Region', fontsize = 14, fontweight = 'bold')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.grid(axis = 'y', linestyle = '--', alpha = 0.5)
plt.tight_layout()
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, # x-position
             height + 20,                     # y-position
             f'{height}',                     # text label
             ha='center',                     # horizontal alignment
             va='bottom',                     # vertical alignment
             fontsize=10,                     
             fontweight = 'bold')
plt.show()
```

```python
#Profit by Region
total_profit = data.groupby('Region')['Profit'].sum().reset_index()
colordict = {'Central': 'red', 'East': 'yellow', 'South': 'blue', 'West': 'orange'}
colors = total_profit['Region'].map(colordict)
plt.figure(figsize = (8, 6))
bars = plt.bar(total_profit['Region'], total_profit['Profit'], 
        color = colors, 
        width = 0.8,
        edgecolor='black', 
        linewidth=1.5)
plt.title('Profit by Region', fontsize = 14, fontweight = 'bold')
plt.xlabel('Region')
plt.ylabel('Profit')
plt.xticks(rotation=45)
plt.grid(axis = 'y', linestyle = '--', alpha = 0.5)
plt.tight_layout()
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2,
             height + 20,
             f'{height}',
             ha = 'center',
             va = 'bottom',
             fontsize = 10,
             fontweight = 'bold')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    'Region': ['North', 'South', 'East', 'West'],
    'Sales': [2200, 1750, 1250, 950]
}
total_sales = pd.DataFrame(data)

# Color dictionary
colordict = {'North': 'skyblue', 'South': 'lightgreen', 'East': 'orange', 'West': 'salmon'}
colors = total_sales['Region'].map(colordict)

# Plot
plt.figure(figsize=(8, 5))
bars = plt.barh(
    total_sales['Region'],
    total_sales['Sales'],
    color=colors,
    edgecolor='black',     #  Adds bar edges
    linewidth=1.5
)

plt.title('Total Sales by Region (Horizontal)', fontsize=14, fontweight='bold')
plt.xlabel('Total Sales')
plt.ylabel('Region')
plt.grid(axis='x', linestyle='--', alpha=0.5)  # grid on x-axis now (horizontal bars)

# Add value labels next to bars
for bar in bars:
    width = bar.get_width()
    plt.text(width + 50, bar.get_y() + bar.get_height()/2, f'{width}',
             ha='left', va='center', fontsize=10)

plt.tight_layout()
plt.show()
```
