```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv(r'C:\DATA ANALYSIS AND DATA SCIENCE PROJECTS\US Superstore Data Visualizations\data\us_superstore_dataset.csv')
```

# BAR GRAPH
*Each bar is a category, bar length is a value*

```python
data.columns
```

```python
#Sales by Region
total_sales = data.groupby('Region')['Sales'].sum().reset_index()
colordict = {'Central': 'red', 'East': 'yellow', 'South': 'blue', 'West': 'orange'}
colors = total_sales['Region'].map(colordict)
plt.figure(figsize = (8, 6))
bars = plt.bar(total_sales['Region'], total_sales['Sales'], 
        color = colors, 
        width = 0.8,
        edgecolor='black', 
        linewidth=1.5)
plt.title('Sales by Region', fontsize = 14, fontweight = 'bold')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.grid(axis = 'y', linestyle = '--', alpha = 0.5)
plt.tight_layout()
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, # x-position
             height + 20,                     # y-position
             f'{height}',                     # text label
             ha='center',                     # horizontal alignment
             va='bottom',                     # vertical alignment
             fontsize=10,                     
             fontweight = 'bold')
plt.show()
```

```python
#Profit by Region
total_profit = data.groupby('Region')['Profit'].sum().reset_index()
colordict = {'Central': 'red', 'East': 'yellow', 'South': 'blue', 'West': 'orange'}
colors = total_profit['Region'].map(colordict)
plt.figure(figsize = (8, 6))
bars = plt.bar(total_profit['Region'], total_profit['Profit'], 
        color = colors, 
        width = 0.8,
        edgecolor='black', 
        linewidth=1.5)
plt.title('Profit by Region', fontsize = 14, fontweight = 'bold')
plt.xlabel('Region')
plt.ylabel('Profit')
plt.xticks(rotation=45)
plt.grid(axis = 'y', linestyle = '--', alpha = 0.5)
plt.tight_layout()
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2,
             height + 20,
             f'{height}',
             ha = 'center',
             va = 'bottom',
             fontsize = 10,
             fontweight = 'bold')
plt.show()
```

```python
#Total Sales by Region (Horizontal)
total_sales = data.groupby('Region')['Sales'].sum().reset_index()
colordict = {'Central': 'skyblue', 'South': 'lightgreen', 'East': 'orange', 'West': 'salmon'}
colors = total_sales['Region'].map(colordict)
plt.figure(figsize=(8, 5))
bars = plt.barh(
    total_sales['Region'],
    total_sales['Sales'],
    color=colors,
    edgecolor='black',     
    linewidth=0.8
)

plt.title('Total Sales by Region (Horizontal)', fontsize=14, fontweight='bold')
plt.xlabel('Total Sales')
plt.ylabel('Region')
plt.grid(axis='x', linestyle='--', alpha=0.5)  

for bar in bars:
    width = bar.get_width()
    plt.text(width + 100, bar.get_y() + bar.get_height()/2, f'{width}',
             ha='left', va='center', fontsize=10)

plt.tight_layout()
plt.show()
```

# GROUPED BAR CHART
*Bars grouped by main category; each group has multiple subcategories*

```python
import numpy as np
#Sales vs Profit by Region
region = data['Region'].unique().tolist()
t_sales = data.groupby('Region')['Sales'].sum().reset_index()
t_profit = data.groupby('Region')['Profit'].sum().reset_index()
sales = t_sales['Sales']
profit = t_profit['Profit']

x = np.arange(len(region))
width = 0.35

plt.bar(x-width/2, 
        sales, 
        width, 
        label = 'Sales',
        color = 'steelblue',
        edgecolor = 'black',
        )
plt.bar(x+width/2, 
        profit, 
        width, 
        label = 'Profit',
        color = 'orange',
        edgecolor = 'black',
        )

plt.xticks(x, region, rotation = 45)
plt.title('Sales vs Profit by Region', fontsize=14, fontweight='bold')
plt.xlabel('Total Sales')
plt.ylabel('Region')
plt.grid(True, linestyle='--', alpha=0.5) 
plt.legend(title = 'Legend')
plt.tight_layout()
plt.show()
```
