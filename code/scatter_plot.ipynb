```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv(r'C:\DATA ANALYSIS AND DATA SCIENCE PROJECTS\US Superstore Data Visualizations\data\us_superstore_dataset.csv')
```

# SCATTER PLOT
*Used to show relationship between two numeric variables*

```python
data['Order Date'] = pd.to_datetime(data['Order Date'], format='%d-%m-%Y', errors='coerce')
data['Ship Date'] = pd.to_datetime(data['Ship Date'], format='%d-%m-%Y', errors='coerce')
data['Shipment Delay (Days)'] = (data['Ship Date'] - data['Order Date']).dt.days
```

```python
#Shipment Delay vs Quantity
plt.figure(figsize = (8, 6))
plt.scatter(data['Quantity'], data['Shipment Delay (Days)'])
plt.xlabel('Quantity', fontsize=12, color='darkblue')
plt.ylabel('Shipment Delay (Days)', fontsize=12, color='darkblue')
plt.title('Shipment Delay vs Quantity', fontsize=14, fontweight='bold')
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
plt.tight_layout()  
plt.legend()
plt.show()
```

```python
data[['Quantity', 'Shipment Delay (Days)']].corr()
```

* **Observation:** There is no visible relationship between shipment delay and quantity.

* **Insight:** The delay in shipment appears to be independent of order size, indicating that shipping processes are likely standardized across different order volumes.

```python
#Sales vs Profit
plt.figure(figsize = (8, 6))
plt.scatter(data['Sales'], data['Profit'], 
            c= 'purple', 
            edgecolors= 'black', 
            linewidths= 0.6)
plt.xlabel('Sales', fontsize = 12)
plt.ylabel('Profit', fontsize = 12)
plt.title('Sales vs Profit', fontsize = 14, fontweight = 'bold')
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)
plt.tight_layout()  
plt.legend()
plt.show()
```
